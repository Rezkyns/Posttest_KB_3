```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
df
```

# MENAMPILKAN INFORMASI DARI BERAPA JUMLAH

```python
df['Country'].value_counts()
```

# GRAFIK BATANG

```python
# Dengan Matplotlib
#plt.figure(figsize=(50,5))
x = list(df["Segment"])
y = list(df["Sales"])

plt.xlabel("customer")
plt.ylabel("penjualan")
plt.title("Grafik Batang")
plt.bar(x,y,color="red")
plt.show()
```

```python
#Dengan Boxplot

plt.title('Grafik Boxplot')
sns.boxplot(data=df,x="Country",y="Sales")
plt.show()
```

# GRAFIK GARIS DAN FLOT

```python
x = list(df["Ship Date"])
y = list(df["Sales"])
plt.title("Grafik Garis")
plt.plot(x,y)
plt.show()
```

# GRAFIK PIE

```python
data_pie = df["Country"].value_counts()
data_pie.head()
```

```python
data_pie = df["Segment"].value_counts().rename_axis("Country")
data_pie.head()
```

```python
data_pie = df["Region"].value_counts().rename_axis("Region").reset_index(name="id_count")
data_pie.head()
```

```python
plt.title("Grafik Pie")
plt.pie(data_pie["id_count"],labels=data_pie["Region"],startangle=90,autopct="%.2f%%",shadow=True,explode=(0,0,0,0.05))
plt.show()
```
