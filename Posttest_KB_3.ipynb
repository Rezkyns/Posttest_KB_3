# Untuk Menampilkan Data

```python
df = pd.read_csv('Car details v3.csv')
df
```

# Untuk Menampilkan Grafik Batang

```python
# Matplotlib
#plt.figure(figsize=(50,5))
x = list(df["fuel"])
y = list(df["seats"])

plt.xlabel("owner")
plt.ylabel("details")
plt.title("Grafik Batang")
plt.bar(x,y,color="blue")
plt.show()
```

# Untuk Menampilkan Grafik garis

```python
x = list(df["year"])
y = list(df["seats"])
plt.title("Grafik Garis")
plt.plot(x,y)
plt.show()
```

# Untuk Menampilkan Grafik Boxplot

```python
# Boxplot

plt.title('Grafik Boxplot')
sns.boxplot(data=df,x="transmission",y="seats")
plt.show()
```

# Pie Chart

```python
data_pie = df["transmission"].value_counts()
data_pie.head()
```

```python
data_pie = df["fuel"].value_counts().rename_axis("transmission")
data_pie.head()
```

```python
data_pie = df["engine"].value_counts().rename_axis("engine").reset_index(name="id_count")
data_pie.head()
```

# Untuk Menampilkan Grafik Pie

```python
plt.title("Grafik Pie")
plt.pie(data_pie["id_count"],labels=data_pie["engine"],startangle=90,autopct="%.2f%%")
plt.show()
```
